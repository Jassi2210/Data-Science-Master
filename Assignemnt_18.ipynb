Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans1- Flask is a lightweight and flexible web framework for Python. It provides the necessary tools and libraries to build web applications in a simple and efficient way. Flask is based on the Werkzeug WSGI toolkit and the Jinja2 template engine.

Some advantages of using Flask are:

1.Lightweight: Flask is a lightweight framework that doesn't require particular tools or libraries, making it easy to use and understand.

2.Flexible: Flask provides a flexible way to create web applications by allowing developers to choose their own tools and libraries to use.

3.Easy to use: Flask is easy to learn and use, and provides a simple and intuitive interface to develop web applications.

4.Modular: Flask is modular, which means it can be extended with various third-party libraries and tools.

5.Scalable: Flask is designed to be scalable, and it can be used to build both small and large web applications.

6.Compatible with different platforms: Flask is compatible with different platforms, including Windows, Mac, and Linux, making it an ideal choice for web development projects.

Q3. What is App routing in Flask? Why do we use app routes?

Ans3- In Flask, app routing is a mechanism for mapping a request URL to a particular view function that will handle the request and generate an HTTP response. App routing in Flask is handled by the @app.route decorator, which is used to specify the URL for a particular view function.

We use app routes in Flask to handle incoming HTTP requests. By mapping URLs to view functions, we can define the behavior of our Flask application and determine what content is displayed to users when they navigate to specific pages on the site. App routing is an essential component of web development with Flask as it allows us to create a dynamic and interactive user experience.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans5- In Flask, the url_for() function is used for URL building. This function takes the name of the endpoint as the first argument, and any additional keyword arguments are used as variables to build the URL.


```python
# Here is a simple example of how to use the url_for() function in Flask:

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the homepage'

@app.route('/user/<username>')
def profile(username):
    return f'This is the profile page for {username}'

with app.test_request_context():
    print(url_for('index')) # Output: '/'
    print(url_for('profile', username='John')) # Output: '/user/John'
```
